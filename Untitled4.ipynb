```python
# ============================================
# DSML Phase 1 - Updated Full Modelling Code
# Cleaned, report-oriented version
# Includes:
# 1) Data loading and cleaning
# 2) Feature engineering
# 3) EDA plots for report
# 4) Feature selection
# 5) Pipeline + model tuning
# 6) Model comparison
# 7) Final model evaluation
# 8) Threshold tuning
# 9) Logistic regression interpretability
# 10) Export tables for report
# ============================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    precision_recall_curve, roc_curve, auc, balanced_accuracy_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# --------------------------------------------
# 0. Basic setup
# --------------------------------------------
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# --------------------------------------------
# 1. Load dataset
# --------------------------------------------
# Update this path if needed
file_path = "DSML Individual Coursework Dataset2526.xlsx"
df = pd.read_excel(file_path)

print("Dataset info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())

# --------------------------------------------
# 2. Initial data understanding
# --------------------------------------------
print("\nMissing values before cleaning:")
print(df.isnull().sum().sort_values(ascending=False))

print("\nDuplicate rows before cleaning:", df.duplicated().sum())

# --------------------------------------------
# 3. Data cleaning
# --------------------------------------------

# 3.1 Remove rows with missing target
df = df.dropna(subset=["Rating"]).copy()

# 3.2 Remove exact duplicates
df = df.drop_duplicates().copy()

# 3.3 Clean Reviews
# Some datasets store Reviews as object due to dirty values, so coerce errors to NaN
df["Reviews_num"] = pd.to_numeric(df["Reviews"], errors="coerce")

# 3.4 Clean Installs
# Example values: "1,000+", "5,000,000+"
df["Installs_num"] = (
    df["Installs"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("+", "", regex=False)
)
df["Installs_num"] = pd.to_numeric(df["Installs_num"], errors="coerce")

# 3.5 Clean Price
# Example values: "$4.99"
df["Price_num"] = (
    df["Price"]
    .astype(str)
    .str.replace("$", "", regex=False)
)
df["Price_num"] = pd.to_numeric(df["Price_num"], errors="coerce")

# 3.6 Clean Size
# Convert size to MB
# Examples: "19M", "14k", "Varies with device"
def convert_size_to_mb(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()

    if x == "Varies with device":
        return np.nan

    if x.endswith("M"):
        try:
            return float(x[:-1])
        except:
            return np.nan

    if x.endswith("k") or x.endswith("K"):
        try:
            return float(x[:-1]) / 1024
        except:
            return np.nan

    try:
        return float(x)
    except:
        return np.nan

df["Size_MB"] = df["Size"].apply(convert_size_to_mb)

# 3.7 Convert Last Updated to datetime
df["Last Updated"] = pd.to_datetime(df["Last Updated"], errors="coerce")

# --------------------------------------------
# 4. Feature engineering
# --------------------------------------------

# 4.1 App age in days
latest_date = df["Last Updated"].max()
df["App_Age_days"] = (latest_date - df["Last Updated"]).dt.days

# 4.2 Free vs Paid
df["is_free"] = (df["Price_num"] == 0).astype(int)

# 4.3 Log transforms for highly skewed variables
df["log_reviews"] = np.log1p(df["Reviews_num"])
df["log_installs"] = np.log1p(df["Installs_num"])

# 4.4 Simplify Genres to primary genre only
df["Genres_primary"] = df["Genres"].astype(str).str.split(";").str[0]

# 4.5 Binary target
df["High_Rating"] = (df["Rating"] >= 4).astype(int)

print("\nDataset shape after cleaning and feature engineering:", df.shape)

print("\nMissing values after cleaning/engineering:")
print(df[[
    "Rating", "Reviews_num", "Installs_num", "Price_num", "Size_MB",
    "Last Updated", "App_Age_days", "log_reviews", "log_installs",
    "Genres_primary", "High_Rating"
]].isnull().sum().sort_values(ascending=False))

# --------------------------------------------
# 5. Tables for report
# --------------------------------------------

# Table 1: Dataset variables summary
dataset_variables_table = pd.DataFrame({
    "Variable": [
        "Category", "Reviews", "Installs", "Size", "Price",
        "Content Rating", "Genres", "Last Updated", "Rating"
    ],
    "Type": [
        "Categorical", "Numeric/Text", "Numeric/Text", "Text", "Text",
        "Categorical", "Categorical", "Date/Text", "Numeric"
    ],
    "Description": [
        "App category", "Number of reviews", "Number of installs", "App size",
        "App price", "Age suitability classification", "App genre(s)",
        "Last update date", "Average user rating"
    ],
    "Used in Model": [
        "Yes", "Log transformed", "Log transformed", "Converted to MB",
        "Converted to numeric", "Yes", "Simplified", "Used for app age", "Target"
    ]
})

# Table 2: Data preprocessing summary
preprocessing_table = pd.DataFrame({
    "Variable": ["Rating", "Reviews", "Installs", "Price", "Size", "Last Updated", "Duplicate rows"],
    "Issue": [
        "Missing target values", "Stored as object / dirty values",
        "Text with commas and plus signs", "Currency string",
        "Mixed units / varies with device", "Text format", "Repeated observations"
    ],
    "Action": [
        "Rows removed", "Converted to numeric",
        "Converted to numeric", "Converted to numeric",
        "Converted to MB", "Converted to datetime", "Removed"
    ]
})

# Table 3: Engineered features
engineered_features_table = pd.DataFrame({
    "New Feature": ["App_Age_days", "is_free", "log_reviews", "log_installs", "Genres_primary", "High_Rating"],
    "Derived From": ["Last Updated", "Price_num", "Reviews_num", "Installs_num", "Genres", "Rating"],
    "Description": [
        "Days since last update",
        "Binary free/paid indicator",
        "Log-transformed review count",
        "Log-transformed install count",
        "Primary genre only",
        "Binary target (1 if Rating >= 4)"
    ]
})

print("\n=== Table 1: Dataset Variables Summary ===")
print(dataset_variables_table)

print("\n=== Table 2: Data Preprocessing Summary ===")
print(preprocessing_table)

print("\n=== Table 3: Engineered Features ===")
print(engineered_features_table)

# --------------------------------------------
# 6. EDA plots for report
# --------------------------------------------

# Figure 1: Original rating distribution
plt.figure(figsize=(6, 4))
sns.histplot(df["Rating"], bins=30)
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Figure 2: Binary target distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="High_Rating", data=df)
plt.title("Binary Rating Distribution")
plt.xlabel("High Rating (0 = <4, 1 = >=4)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Figure 3: Numerical variable distributions
# Use transformed variables for readability
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

sns.histplot(df["log_reviews"].dropna(), bins=30, ax=axes[0, 0])
axes[0, 0].set_title("log_reviews")

sns.histplot(df["log_installs"].dropna(), bins=30, ax=axes[0, 1])
axes[0, 1].set_title("log_installs")

sns.histplot(df["Price_num"].dropna(), bins=30, ax=axes[0, 2])
axes[0, 2].set_title("Price_num")

sns.histplot(df["Size_MB"].dropna(), bins=30, ax=axes[1, 0])
axes[1, 0].set_title("Size_MB")

sns.histplot(df["App_Age_days"].dropna(), bins=30, ax=axes[1, 1])
axes[1, 1].set_title("App_Age_days")

axes[1, 2].axis("off")

plt.suptitle("Distributions of Numerical Variables")
plt.tight_layout()
plt.show()

# Figure 4: Correlation heatmap
corr_cols = ["Rating", "log_reviews", "log_installs", "Price_num", "Size_MB", "App_Age_days"]
corr = df[corr_cols].corr(method="spearman")

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.tight_layout()
plt.show()

# Figure 5: Rating by top categories
top_categories = df["Category"].value_counts().head(10).index
df_top_cat = df[df["Category"].isin(top_categories)].copy()

plt.figure(figsize=(10, 5))
sns.boxplot(x="Category", y="Rating", data=df_top_cat)
plt.xticks(rotation=45)
plt.title("Rating by Top 10 App Categories")
plt.tight_layout()
plt.show()

# Figure 6: Free vs Paid
# More report-friendly labels
df["App_Type_Label"] = np.where(df["is_free"] == 1, "Free", "Paid")

plt.figure(figsize=(6, 4))
sns.boxplot(x="App_Type_Label", y="Rating", data=df)
plt.title("Rating by App Type")
plt.xlabel("Type")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()

# Figure 7: Content Rating vs Rating
plt.figure(figsize=(8, 4))
sns.boxplot(x="Content Rating", y="Rating", data=df)
plt.xticks(rotation=45)
plt.title("Rating by Content Rating")
plt.tight_layout()
plt.show()

# Figure 8: Rating vs log_reviews
plt.figure(figsize=(6, 4))
sns.scatterplot(x="log_reviews", y="Rating", data=df, alpha=0.25)
plt.title("Rating vs log(Reviews)")
plt.tight_layout()
plt.show()

# Figure 9: Rating vs log_installs
plt.figure(figsize=(6, 4))
sns.scatterplot(x="log_installs", y="Rating", data=df, alpha=0.25)
plt.title("Rating vs log(Installs)")
plt.tight_layout()
plt.show()

# --------------------------------------------
# 7. Feature selection for modelling
# --------------------------------------------

# Final selected columns
categorical_features = ["Category", "Type", "Content Rating", "Genres_primary"]
numeric_features = ["Price_num", "Size_MB", "App_Age_days", "is_free", "log_reviews", "log_installs"]

target = "High_Rating"

model_df = df[categorical_features + numeric_features + [target]].copy()

print("\nFinal modelling dataset shape:", model_df.shape)
print("\nTarget distribution:")
print(model_df[target].value_counts())

# --------------------------------------------
# 8. Train-test split
# --------------------------------------------
X = model_df.drop(columns=[target])
y = model_df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\nTrain shape:", X_train.shape)
print("Test shape:", X_test.shape)

# --------------------------------------------
# 9. Preprocessing pipeline
# --------------------------------------------
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

# --------------------------------------------
# 10. Models and parameter grids
# --------------------------------------------
models_and_params = {
    "Logistic Regression": {
        "model": LogisticRegression(max_iter=5000, random_state=42),
        "params": {
            "classifier__C": [0.1, 1, 10]
        }
    },
    "KNN": {
        "model": KNeighborsClassifier(),
        "params": {
            "classifier__n_neighbors": [5, 7, 9, 11],
            "classifier__weights": ["uniform", "distance"]
        }
    },
    "SVM": {
        "model": SVC(probability=True, random_state=42),
        "params": {
            "classifier__C": [0.5, 1, 5],
            "classifier__kernel": ["rbf"]
        }
    },
    "Decision Tree": {
        "model": DecisionTreeClassifier(random_state=42),
        "params": {
            "classifier__max_depth": [3, 5, 7, None],
            "classifier__min_samples_split": [2, 5, 10]
        }
    }
}

# --------------------------------------------
# 11. Model tuning and comparison
# --------------------------------------------
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = []
best_models = {}

for model_name, config in models_and_params.items():
    print(f"\nTuning {model_name}...")

    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("classifier", config["model"])
    ])

    grid = GridSearchCV(
        estimator=pipe,
        param_grid=config["params"],
        scoring="f1",
        cv=cv,
        n_jobs=-1
    )

    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_
    best_models[model_name] = best_model

    y_pred = best_model.predict(X_test)
    y_proba = best_model.predict_proba(X_test)[:, 1]

    result_row = {
        "Model": model_name,
        "Best Params": grid.best_params_,
        "CV F1": grid.best_score_,
        "Test Accuracy": accuracy_score(y_test, y_pred),
        "Test Precision": precision_score(y_test, y_pred),
        "Test Recall": recall_score(y_test, y_pred),
        "Test F1": f1_score(y_test, y_pred),
        "Test Macro F1": f1_score(y_test, y_pred, average="macro"),
        "Test Balanced Accuracy": balanced_accuracy_score(y_test, y_pred),
        "Test ROC AUC": roc_auc_score(y_test, y_proba)
    }

    results.append(result_row)

results_df = pd.DataFrame(results).sort_values(by="Test F1", ascending=False).reset_index(drop=True)

print("\n=== Model Comparison Results ===")
print(results_df)

# --------------------------------------------
# 12. Model comparison plots
# --------------------------------------------

# Figure 10: Test F1 comparison
plt.figure(figsize=(7, 4))
sns.barplot(x="Model", y="Test F1", data=results_df)
plt.title("Model Comparison by Test F1 Score")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()

# Figure 11: Test Macro F1 comparison
plt.figure(figsize=(7, 4))
sns.barplot(x="Model", y="Test Macro F1", data=results_df)
plt.title("Model Comparison by Test Macro F1 Score")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()

# --------------------------------------------
# 13. Select final model (best by Test F1)
# --------------------------------------------
best_model_name = results_df.iloc[0]["Model"]
final_model = best_models[best_model_name]

print(f"\nSelected final model: {best_model_name}")

y_pred_final = final_model.predict(X_test)
y_proba_final = final_model.predict_proba(X_test)[:, 1]

# --------------------------------------------
# 14. Final evaluation
# --------------------------------------------

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_final)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f"Confusion Matrix - {best_model_name}")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.tight_layout()
plt.show()

# ROC curve
fpr, tpr, _ = roc_curve(y_test, y_proba_final)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"Classifier (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Positive label: 1)")
plt.title(f"ROC Curve - {best_model_name}")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# Precision-Recall curve
precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_proba_final)
pr_auc = auc(recall_vals, precision_vals)

plt.figure(figsize=(6, 5))
plt.plot(recall_vals, precision_vals, label=f"PR AUC = {pr_auc:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title(f"Precision-Recall Curve - {best_model_name}")
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

# Classification report table
report_dict = classification_report(y_test, y_pred_final, output_dict=True)
classification_report_df = pd.DataFrame(report_dict).transpose()

print("\n=== Final Classification Report ===")
print(classification_report_df)

# --------------------------------------------
# 15. Threshold tuning for final model
# --------------------------------------------
thresholds = np.arange(0.30, 0.71, 0.02)
threshold_results = []

for thr in thresholds:
    y_pred_thr = (y_proba_final >= thr).astype(int)

    threshold_results.append({
        "threshold": thr,
        "precision": precision_score(y_test, y_pred_thr, zero_division=0),
        "recall": recall_score(y_test, y_pred_thr, zero_division=0),
        "f1": f1_score(y_test, y_pred_thr, zero_division=0)
    })

threshold_df = pd.DataFrame(threshold_results)

plt.figure(figsize=(8, 5))
plt.plot(threshold_df["threshold"], threshold_df["f1"], label="F1")
plt.plot(threshold_df["threshold"], threshold_df["precision"], label="Precision")
plt.plot(threshold_df["threshold"], threshold_df["recall"], label="Recall")
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Threshold Tuning Performance")
plt.legend()
plt.tight_layout()
plt.show()

print("\n=== Threshold Tuning Results ===")
print(threshold_df)

# --------------------------------------------
# 16. Logistic regression interpretability
# --------------------------------------------
# Fit logistic regression again (best tuned version if available), even if not final,
# to support report interpretation with coefficients.

log_reg_model = best_models["Logistic Regression"]

# Get feature names after preprocessing
preprocessor_fitted = log_reg_model.named_steps["preprocessor"]
classifier_fitted = log_reg_model.named_steps["classifier"]

# Numerical feature names
num_feature_names = numeric_features

# Categorical feature names after one-hot encoding
cat_encoder = preprocessor_fitted.named_transformers_["cat"].named_steps["onehot"]
cat_feature_names = cat_encoder.get_feature_names_out(categorical_features)

all_feature_names = np.concatenate([num_feature_names, cat_feature_names])

coefficients = classifier_fitted.coef_[0]

coef_df = pd.DataFrame({
    "feature": all_feature_names,
    "coefficient": coefficients
}).sort_values(by="coefficient", key=lambda x: np.abs(x), ascending=False)

top_coef_df = coef_df.head(15).copy()

plt.figure(figsize=(8, 6))
sns.barplot(data=top_coef_df, x="coefficient", y="feature")
plt.title("Top 15 Logistic Regression Coefficients")
plt.tight_layout()
plt.show()

print("\n=== Top Logistic Regression Coefficients ===")
print(top_coef_df)

# --------------------------------------------
# 17. Export CSVs for report tables
# --------------------------------------------
results_df.to_csv("model_comparison_results.csv", index=False)
threshold_df.to_csv("threshold_tuning_results.csv", index=False)
top_coef_df.to_csv("logistic_coefficients_interpretability.csv", index=False)
classification_report_df.to_csv("final_classification_report_table.csv", index=True)
model_df.to_csv("apps_model_ready.csv", index=False)

print("\nExported files:")
print("- model_comparison_results.csv")
print("- threshold_tuning_results.csv")
print("- logistic_coefficients_interpretability.csv")
print("- final_classification_report_table.csv")
print("- apps_model_ready.csv")

# --------------------------------------------
# 18. Optional: save tables for easy copy into report
# --------------------------------------------
dataset_variables_table.to_csv("table_dataset_variables_summary.csv", index=False)
preprocessing_table.to_csv("table_preprocessing_summary.csv", index=False)
engineered_features_table.to_csv("table_engineered_features.csv", index=False)

print("- table_dataset_variables_summary.csv")
print("- table_preprocessing_summary.csv")
print("- table_engineered_features.csv")
print("\nDone.")
```
