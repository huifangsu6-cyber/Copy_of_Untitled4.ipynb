```python
# ============================================================
# PHASE 1: APP RATING PREDICTION
# Coursework-aligned version (course methods only)
# Methods used:
# - Data cleaning
# - Feature engineering
# - Correlation-based feature selection
# - One-hot encoding
# - RobustScaler
# - Train/test split
# - Logistic Regression
# - KNN
# - SVM
# - Decision Tree
# - GridSearchCV
# - Stratified K-Fold Cross Validation
# - Evaluation: Accuracy, Precision, Recall, F1, ROC-AUC, PR-AUC
# - Visualisation for report
# ============================================================

# =========================
# 1. IMPORT LIBRARIES
# =========================
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import spearmanr

from sklearn.model_selection import (
    train_test_split,
    StratifiedKFold,
    GridSearchCV
)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix, ConfusionMatrixDisplay,
    roc_auc_score, RocCurveDisplay,
    precision_recall_curve, auc
)

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Display settings
pd.set_option("display.max_columns", 200)
sns.set(style="whitegrid", font_scale=1.0)

# =========================
# 2. LOAD DATA
# =========================
file_path = "DSML Individual Coursework Dataset2526.xlsx"

apps = pd.read_excel(file_path, sheet_name="Apps")
dictionary = pd.read_excel(file_path, sheet_name="Dictionary")

print("Apps dataset shape:", apps.shape)
print("\nColumns:")
print(apps.columns.tolist())

print("\nFirst 5 rows:")
display(apps.head())

print("\nDictionary sheet preview:")
display(dictionary.head(20))

# =========================
# 3. INITIAL DATA UNDERSTANDING
# =========================
print("\nDataset info:")
apps.info()

print("\nMissing values:")
display(apps.isnull().sum().sort_values(ascending=False).to_frame("missing_count"))

print("\nDuplicate full rows:", apps.duplicated().sum())
print("Duplicate app names:", apps.duplicated(subset=["App"]).sum())

# =========================
# 4. DATA CLEANING
# Why:
# Raw columns such as Reviews, Installs, Price and Size are not
# in a model-ready numeric format.
# =========================
df = apps.copy()

# Remove exact duplicate rows
df = df.drop_duplicates()

# ---------- Helper functions ----------
def clean_reviews(x):
    return pd.to_numeric(x, errors="coerce")

def clean_installs(x):
    if pd.isna(x):
        return np.nan
    s = str(x).replace(",", "").replace("+", "").strip()
    return pd.to_numeric(s, errors="coerce")

def clean_price(x):
    if pd.isna(x):
        return np.nan
    s = str(x).replace("$", "").strip()
    return pd.to_numeric(s, errors="coerce")

def clean_size_to_mb(x):
    """
    Convert app size to MB:
    19M -> 19
    512k -> 0.5 approx
    Varies with device -> NaN
    """
    if pd.isna(x):
        return np.nan

    s = str(x).strip()

    if s.lower() == "varies with device":
        return np.nan

    s = s.replace(",", "")
    match = re.match(r"^([\d\.]+)\s*([kKmM]?)$", s)

    if match:
        value = float(match.group(1))
        unit = match.group(2).lower()

        if unit == "m":
            return value
        elif unit == "k":
            return value / 1024.0
        else:
            return value

    return pd.to_numeric(s, errors="coerce")

# ---------- Apply cleaning ----------
df["Reviews_num"] = df["Reviews"].apply(clean_reviews)
df["Installs_num"] = df["Installs"].apply(clean_installs)
df["Price_num"] = df["Price"].apply(clean_price)
df["Size_MB"] = df["Size"].apply(clean_size_to_mb)
df["Last_Updated_dt"] = pd.to_datetime(df["Last Updated"], errors="coerce")

# Remove rows with missing target
df = df[df["Rating"].notna()].copy()

# Keep only plausible rating values
df = df[(df["Rating"] >= 0) & (df["Rating"] <= 5)].copy()

print("\nShape after cleaning basic rows:", df.shape)

# =========================
# 5. TARGET DESIGN
# Why:
# Business objective is better aligned with identifying apps likely
# to achieve strong user satisfaction rather than predicting an exact decimal.
# Binary target:
# good_rating = 1 if Rating >= 4, else 0
# =========================
df["good_rating"] = np.where(df["Rating"] >= 4.0, 1, 0)

print("\nTarget distribution (counts):")
display(df["good_rating"].value_counts().to_frame("count"))

print("\nTarget distribution (%):")
display((df["good_rating"].value_counts(normalize=True) * 100).round(2).to_frame("percentage"))

# =========================
# 6. FEATURE ENGINEERING
# =========================
reference_date = df["Last_Updated_dt"].max()
df["App_Age_days"] = (reference_date - df["Last_Updated_dt"]).dt.days

df["is_free"] = np.where(df["Price_num"] == 0, 1, 0)

# log transforms help reduce skewness
df["log_reviews"] = np.log1p(df["Reviews_num"])
df["log_installs"] = np.log1p(df["Installs_num"])

# Keep first genre only to reduce complexity slightly
df["Genres_primary"] = df["Genres"].astype(str).str.split(";").str[0]

print("\nEngineered feature preview:")
display(df[[
    "Rating", "good_rating", "Reviews_num", "Installs_num",
    "Price_num", "Size_MB", "App_Age_days", "is_free",
    "log_reviews", "log_installs", "Genres_primary"
]].head())

# =========================
# 7. DATA VISUALISATION FOR REPORT
# =========================

# 7.1 Original rating distribution
plt.figure(figsize=(7, 4))
sns.histplot(df["Rating"], bins=20, kde=False)
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# 7.2 Binary target distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="good_rating", data=df)
plt.title("Distribution of Binary Target")
plt.xlabel("good_rating (0 = <4, 1 = >=4)")
plt.ylabel("Count")
plt.show()

# 7.3 Histograms of numeric variables
numeric_plot_cols = ["Reviews_num", "Installs_num", "Price_num", "Size_MB", "App_Age_days"]
df[numeric_plot_cols].hist(figsize=(14, 8), bins=30)
plt.suptitle("Distributions of Numerical Variables", y=1.02)
plt.tight_layout()
plt.show()

# =========================
# 8. CORRELATION-BASED FEATURE SELECTION
# Only for numeric variables
# =========================
numeric_candidates = [
    "Reviews_num", "Installs_num", "Price_num",
    "Size_MB", "App_Age_days", "log_reviews", "log_installs"
]

corr_results = []

for col in numeric_candidates:
    valid = df[[col, "Rating"]].dropna()
    corr, pval = spearmanr(valid[col], valid["Rating"])
    corr_results.append({
        "feature": col,
        "spearman_corr_with_rating": corr,
        "p_value": pval
    })

corr_df = pd.DataFrame(corr_results).sort_values(
    by="spearman_corr_with_rating",
    key=lambda s: s.abs(),
    ascending=False
)

print("\nSpearman correlations with Rating:")
display(corr_df)

# Correlation heatmap
heatmap_cols = ["Rating"] + numeric_candidates
corr_matrix = df[heatmap_cols].corr(numeric_only=True)

plt.figure(figsize=(9, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap for Numerical Variables")
plt.show()

# =========================
# 9. RELATIONSHIP ANALYSIS FOR CATEGORICAL VARIABLES
# Correlation is not appropriate for nominal categories,
# so we compare group means and distributions.
# =========================

# 9.1 Category summary
category_summary = (
    df.groupby("Category")
      .agg(mean_rating=("Rating", "mean"),
           app_count=("App", "count"))
      .sort_values("mean_rating", ascending=False)
)

print("\nTop categories by mean rating:")
display(category_summary.head(10))

print("\nBottom categories by mean rating:")
display(category_summary.tail(10))

# 9.2 Type summary
type_summary = (
    df.groupby("Type")
      .agg(mean_rating=("Rating", "mean"),
           app_count=("App", "count"))
      .sort_values("mean_rating", ascending=False)
)

print("\nRating by Type:")
display(type_summary)

# 9.3 Content Rating summary
content_summary = (
    df.groupby("Content Rating")
      .agg(mean_rating=("Rating", "mean"),
           app_count=("App", "count"))
      .sort_values("mean_rating", ascending=False)
)

print("\nRating by Content Rating:")
display(content_summary)

# 9.4 Boxplots
top_categories = df["Category"].value_counts().head(10).index

plt.figure(figsize=(12, 5))
sns.boxplot(data=df[df["Category"].isin(top_categories)], x="Category", y="Rating")
plt.title("Rating by Top 10 App Categories")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x="Type", y="Rating")
plt.title("Rating by App Type")
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x="Content Rating", y="Rating")
plt.title("Rating by Content Rating")
plt.xticks(rotation=45)
plt.show()

# 9.5 Scatterplots
sample_df = df.sample(min(len(df), 3000), random_state=42)

plt.figure(figsize=(6, 4))
sns.scatterplot(data=sample_df, x="log_reviews", y="Rating", alpha=0.5)
plt.title("Rating vs log(Reviews)")
plt.show()

plt.figure(figsize=(6, 4))
sns.scatterplot(data=sample_df, x="log_installs", y="Rating", alpha=0.5)
plt.title("Rating vs log(Installs)")
plt.show()

# =========================
# 10. FINAL FEATURE SET
# Based on:
# - business meaning
# - course-taught feature engineering
# - correlation / relationship evidence
# =========================
feature_cols = [
    "Category",
    "Type",
    "Content Rating",
    "Genres_primary",
    "Reviews_num",
    "Installs_num",
    "Price_num",
    "Size_MB",
    "App_Age_days",
    "is_free",
    "log_reviews",
    "log_installs"
]

target_col = "good_rating"

model_df = df[feature_cols + [target_col]].copy()

print("\nFinal modelling dataframe shape:", model_df.shape)

print("\nMissing values in final modelling dataframe:")
display(model_df.isnull().sum().sort_values(ascending=False).to_frame("missing_count"))

# =========================
# 11. TRAIN / TEST SPLIT
# Stratified split preserves class distribution.
# =========================
X = model_df[feature_cols]
y = model_df[target_col]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

print("\nTrain target distribution (%):")
display((y_train.value_counts(normalize=True) * 100).round(2).to_frame("percentage"))

print("\nTest target distribution (%):")
display((y_test.value_counts(normalize=True) * 100).round(2).to_frame("percentage"))

# =========================
# 12. PREPROCESSING PIPELINE
# - Median imputation for numerical features
# - Most frequent imputation for categorical features
# - OneHotEncoder for categorical variables
# - RobustScaler for numerical variables
# Important: done inside pipeline to avoid data leakage
# =========================
categorical_features = ["Category", "Type", "Content Rating", "Genres_primary"]
numerical_features = [
    "Reviews_num", "Installs_num", "Price_num",
    "Size_MB", "App_Age_days", "is_free",
    "log_reviews", "log_installs"
]

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

# =========================
# 13. MODEL SETUP
# Course methods only:
# - Logistic Regression
# - KNN
# - SVM
# - Decision Tree
# =========================
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

pipelines = {
    "Logistic Regression": Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", LogisticRegression(max_iter=2000, class_weight="balanced"))
    ]),

    "KNN": Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", KNeighborsClassifier())
    ]),

    "SVM": Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", SVC(probability=True, class_weight="balanced"))
    ]),

    "Decision Tree": Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", DecisionTreeClassifier(class_weight="balanced", random_state=42))
    ])
}

param_grids = {
    "Logistic Regression": {
        "model__C": [0.01, 0.1, 1, 10]
    },
    "KNN": {
        "model__n_neighbors": [5, 9, 15],
        "model__weights": ["uniform", "distance"],
        "model__p": [1, 2]
    },
    "SVM": {
        "model__C": [0.1, 1, 10],
        "model__kernel": ["linear", "rbf"]
    },
    "Decision Tree": {
        "model__max_depth": [3, 5, 8, None],
        "model__min_samples_split": [2, 5, 10],
        "model__min_samples_leaf": [1, 2, 4]
    }
}

# =========================
# 14. MODEL TUNING + CROSS VALIDATION
# Refit metric = F1 because accuracy alone may be misleading
# under class imbalance.
# =========================
results = []
best_estimators = {}

for model_name, pipe in pipelines.items():
    print(f"\n{'='*70}")
    print(f"Tuning {model_name}")
    print(f"{'='*70}")

    grid = GridSearchCV(
        estimator=pipe,
        param_grid=param_grids[model_name],
        scoring={
            "accuracy": "accuracy",
            "precision": "precision",
            "recall": "recall",
            "f1": "f1",
            "roc_auc": "roc_auc"
        },
        refit="f1",
        cv=cv,
        n_jobs=-1,
        verbose=1
    )

    grid.fit(X_train, y_train)

    best_estimators[model_name] = grid.best_estimator_

    y_pred = grid.best_estimator_.predict(X_test)
    y_proba = grid.best_estimator_.predict_proba(X_test)[:, 1]

    result_row = {
        "Model": model_name,
        "Best Params": grid.best_params_,
        "CV Accuracy": grid.cv_results_["mean_test_accuracy"][grid.best_index_],
        "CV Precision": grid.cv_results_["mean_test_precision"][grid.best_index_],
        "CV Recall": grid.cv_results_["mean_test_recall"][grid.best_index_],
        "CV F1": grid.cv_results_["mean_test_f1"][grid.best_index_],
        "CV ROC AUC": grid.cv_results_["mean_test_roc_auc"][grid.best_index_],
        "Test Accuracy": accuracy_score(y_test, y_pred),
        "Test Precision": precision_score(y_test, y_pred),
        "Test Recall": recall_score(y_test, y_pred),
        "Test F1": f1_score(y_test, y_pred),
        "Test ROC AUC": roc_auc_score(y_test, y_proba)
    }

    results.append(result_row)

results_df = pd.DataFrame(results).sort_values(by="Test F1", ascending=False)

print("\n================ MODEL COMPARISON ================")
display(results_df)

# Optional model comparison plot
plt.figure(figsize=(8, 4))
sns.barplot(data=results_df, x="Model", y="Test F1")
plt.title("Model Comparison by Test F1 Score")
plt.xticks(rotation=20)
plt.show()

# =========================
# 15. SELECT BEST MODEL
# =========================
best_model_name = results_df.iloc[0]["Model"]
best_model = best_estimators[best_model_name]

print(f"\nBest model selected: {best_model_name}")

# =========================
# 16. FINAL EVALUATION
# =========================
y_pred_best = best_model.predict(X_test)
y_proba_best = best_model.predict_proba(X_test)[:, 1]

print("\nClassification report:")
print(classification_report(y_test, y_pred_best))

# 16.1 Confusion Matrix
cm = confusion_matrix(y_test, y_pred_best)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title(f"Confusion Matrix - {best_model_name}")
plt.show()

# 16.2 ROC Curve
RocCurveDisplay.from_predictions(y_test, y_proba_best)
plt.title(f"ROC Curve - {best_model_name}")
plt.show()

# 16.3 Precision-Recall Curve
precision_vals, recall_vals, thresholds = precision_recall_curve(y_test, y_proba_best)
pr_auc = auc(recall_vals, precision_vals)

plt.figure(figsize=(6, 5))
plt.plot(recall_vals, precision_vals, label=f"PR AUC = {pr_auc:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title(f"Precision-Recall Curve - {best_model_name}")
plt.legend()
plt.show()

print(f"PR AUC: {pr_auc:.4f}")

# =========================
# 17. THRESHOLD TUNING
# Why:
# Default threshold 0.50 may not be optimal.
# This is useful for business interpretation and report discussion.
# =========================
threshold_candidates = np.arange(0.30, 0.71, 0.02)

threshold_results = []
for t in threshold_candidates:
    y_pred_t = (y_proba_best >= t).astype(int)
    threshold_results.append({
        "threshold": t,
        "precision": precision_score(y_test, y_pred_t, zero_division=0),
        "recall": recall_score(y_test, y_pred_t, zero_division=0),
        "f1": f1_score(y_test, y_pred_t, zero_division=0),
        "accuracy": accuracy_score(y_test, y_pred_t)
    })

threshold_df = pd.DataFrame(threshold_results).sort_values(by="f1", ascending=False)

print("\nTop threshold results by F1:")
display(threshold_df.head(10))

best_threshold = threshold_df.iloc[0]["threshold"]
print(f"\nBest threshold by F1: {best_threshold:.2f}")

plt.figure(figsize=(8, 4))
sns.lineplot(data=threshold_df.sort_values("threshold"), x="threshold", y="f1", label="F1")
sns.lineplot(data=threshold_df.sort_values("threshold"), x="threshold", y="precision", label="Precision")
sns.lineplot(data=threshold_df.sort_values("threshold"), x="threshold", y="recall", label="Recall")
plt.title("Threshold Tuning Performance")
plt.ylabel("Score")
plt.show()

# =========================
# 18. INTERPRETABILITY
# For coursework-friendly interpretation:
# - If Logistic Regression is best, inspect coefficients
# - If Decision Tree is best, inspect simple tree importance
# =========================
feature_names = best_model.named_steps["preprocessor"].get_feature_names_out()

if best_model_name == "Logistic Regression":
    coefs = best_model.named_steps["model"].coef_[0]
    coef_df = pd.DataFrame({
        "feature": feature_names,
        "coefficient": coefs,
        "abs_coefficient": np.abs(coefs)
    }).sort_values(by="abs_coefficient", ascending=False)

    print("\nTop 20 Logistic Regression coefficients:")
    display(coef_df.head(20))

    plt.figure(figsize=(10, 8))
    sns.barplot(data=coef_df.head(15), x="coefficient", y="feature")
    plt.title("Top 15 Logistic Regression Coefficients")
    plt.show()

elif best_model_name == "Decision Tree":
    importances = best_model.named_steps["model"].feature_importances_
    importance_df = pd.DataFrame({
        "feature": feature_names,
        "importance": importances
    }).sort_values(by="importance", ascending=False)

    print("\nTop 20 Decision Tree feature importances:")
    display(importance_df.head(20))

    plt.figure(figsize=(10, 8))
    sns.barplot(data=importance_df.head(15), x="importance", y="feature")
    plt.title("Top 15 Decision Tree Feature Importances")
    plt.show()

else:
    print(f"\nThe best model is {best_model_name}.")
    print("Interpretability will mainly rely on model comparison, confusion matrix, ROC/PR curves,")
    print("threshold analysis, and feature-target relationship plots from EDA.")

# =========================
# 19. SAVE OUTPUT FILES
# =========================
model_df.to_csv("apps_model_ready.csv", index=False)
results_df.to_csv("model_comparison_results.csv", index=False)
threshold_df.to_csv("threshold_tuning_results.csv", index=False)

print("\nSaved files:")
print("- apps_model_ready.csv")
print("- model_comparison_results.csv")
print("- threshold_tuning_results.csv")
```
